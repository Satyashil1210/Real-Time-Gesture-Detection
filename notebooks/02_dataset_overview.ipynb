# Dataset Overview

This notebook inspects the dataset present in `data/raw` folder
and shows sample distribution and images.


```python
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
DATA_DIR = Path("../data/raw")

if not DATA_DIR.exists():
    print("data/raw folder not found.")
else:
    for gesture_dir in DATA_DIR.iterdir():
        if gesture_dir.is_dir():
            count = len(list(gesture_dir.glob("*.jpg")))
            print(f"{gesture_dir.name}: {count} images")
```

```python
gesture_dirs = [d for d in DATA_DIR.iterdir() if d.is_dir()]

if not gesture_dirs:
    print("No dataset yet.")
else:
    chosen = random.choice(gesture_dirs)
    img_file = random.choice(list(chosen.glob("*.jpg")))

    img = Image.open(img_file)
    plt.imshow(img)
    plt.title(f"Sample from {chosen.name}")
    plt.axis("off")
    plt.show()
```
